```python
data = data.map(lambda x,y: (x/255, y))
```

```python
data.as_numpy_iterator().next()
```

```python
train_size = int(len(data)*.7)
val_size = int(len(data)*.2)
test_size = int(len(data)*.1)
```

```python
train_size
```

```python
train = data.take(train_size)
val = data.skip(train_size).take(val_size)
test = data.skip(train_size+val_size).take(test_size)
```

```python

train
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout
```

```python
model = Sequential()
```

```python
model.add(Conv2D(16, (3,3), 1, activation='relu', input_shape=(256,256,3)))
model.add(MaxPooling2D())
model.add(Conv2D(32, (3,3), 1, activation='relu'))
model.add(MaxPooling2D())
model.add(Conv2D(16, (3,3), 1, activation='relu'))
model.add(MaxPooling2D())
model.add(Flatten())
model.add(Dense(256, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile('adam', loss=tf.losses.BinaryCrossentropy(), metrics=['accuracy'])
```

```python
model.summary()
```

```python
logdir='logs'
```

```python

tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=logdir)
```

```python
hist = model.fit(train, epochs=3, validation_data=val, callbacks=[tensorboard_callback])
```

```python
fig = plt.figure()
plt.plot(hist.history['loss'], color='teal', label='loss')
plt.plot(hist.history['val_loss'], color='orange', label='val_loss')
fig.suptitle('Loss', fontsize=20)
plt.legend(loc="upper left")
plt.show()
```

```python
fig = plt.figure()
plt.plot(hist.history['accuracy'], color='teal', label='accuracy')
plt.plot(hist.history['val_accuracy'], color='orange', label='val_accuracy')
fig.suptitle('Accuracy', fontsize=20)
plt.legend(loc="upper left")
plt.show()
```

```python
from tensorflow.keras.metrics import Precision, Recall, BinaryAccuracy
```

```python
pre = Precision()
re = Recall()
acc = BinaryAccuracy()
```

```python
for batch in test.as_numpy_iterator():
    X, y = batch
    yhat = model.predict(X)
    pre.update_state(y, yhat)
    re.update_state(y, yhat)
    acc.update_state(y, yhat)
```

```python
print(pre.result(), re.result(), acc.result())
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
from PIL import Image


y_true = [0, 1, 0, 1, 1, 1, 0, 0, 1, 0]
y_pred = [0, 1, 1, 1, 1, 0, 0, 1, 1, 0]


cm = confusion_matrix(y_true, y_pred)


image_paths = [
    "/content/sample_data/data/datat1/04_02_0001.png",
    "/content/sample_data/data/data2/augmented_9_6.png",
    "/content/sample_data/data/data3/02_0001.png",
    "/content/sample_data/data/data6/06_02_0001.png",
    "/content/sample_data/data/data10/01_01_0005.png"
]


images = [Image.open(path) for path in image_paths]


plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, cmap="Reds", xticklabels=["Negative", "Positive"], yticklabels=["Negative", "Positive" ], fmt="f")


for i in range(len(images)):
    for j in range(len(images)):
        plt.gca().add_patch(plt.Rectangle((i, j), 1, 1, fill=False, edgecolor='red', lw=1.5))
        plt.gca().annotate('', xy=(i+0.5, j+0.5), xytext=(0, 0), textcoords='offset points', arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0'))
        plt.gca().imshow(images[i], extent=(i, i+1, j, j+1), aspect='auto', origin='upper')

plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.title('Confusion Matrix')
plt.show()
```
